```python
import pandas as pd
dataset = pd.read_csv("/home/tintin/Study/Data-Science/Statistical-Machine-Learning/Hierarchical-Clustering/Datasets/Mall_Customers.csv")
dataset
```

```python
X = dataset.iloc[:,[3,4]].values
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], color = "red")
plt.xlabel("Annual Income(K$)")
plt.ylabel("Spending Score")
plt.title("Annual Income vs Spending Score")
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = "ward"))
plt.xlabel("Customers")
plt.ylabel("Euclidean distances")
plt.title("Dendrogram")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity= "euclidean", linkage = "ward")
y_hc = hc.fit_predict(X)
```

```python
from matplotlib.colors import ListedColormap
raw_colors = ("red", "green", "black", "cyan", "magenta")
colors = ListedColormap(raw_colors)
for i in range(5):
    plt.scatter(X[y_hc == i, 0], X[y_hc == i, 1], s = 100, c = colors(i), label = raw_colors[i])
plt.xlabel("Annual Income (k$)")
plt.ylabel("Speding Score(1-100)")
plt.legend(loc='right')
plt.title("Customers Clustering")
plt.show()
```
