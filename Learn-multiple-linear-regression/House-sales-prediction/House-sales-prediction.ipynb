```python
#Importing the libraries
import numpy as np
import pandas as pd
from math import sqrt
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
dataset_train = pd.read_csv("/home/tintin/Study/Data-science/Statistical-machine-learning/Learn-multiple-linear-regression/Datasets/kc_house_Train.csv") #If you use Linux operation system, you shouldn't put sign "/" before /path folder
dataset_test = pd.read_csv("/home/tintin/Study/Data-science/Statistical-machine-learning/Learn-multiple-linear-regression/Datasets/kc_house_Test.csv")
dataset_train
```

```python
X_train = dataset_train.iloc[:, 4:9].values
Y_train = dataset_train.iloc[:, 3].values

X_test = dataset_test.iloc[:, 4:9].values
Y_test = dataset_test.iloc[:, 3].values

X_train = np.append(X_train, dataset_train.iloc[:,13:15].values,axis=1)
X_test = np.append(X_test, dataset_test.iloc[:,13:15].values,axis=1)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, Y_train)
```

```python
Y_test_pred = lin_reg.predict(X_test)
Y_train_pred = lin_reg.predict(X_train)

print("R2 score training = ",lin_reg.score(X_train, Y_train))
print("RMSE score training =  ", sqrt(mean_squared_error(Y_train, Y_train_pred)))
print("MSE score training = ", mean_squared_error(Y_train, Y_train_pred))
print("MAE score training = ", mean_absolute_error(Y_train, Y_train_pred))

print("R2 score testing = ",lin_reg.score(X_test, Y_test))
print("RMSE score testing = ", sqrt(mean_squared_error(Y_test, Y_test_pred)))
print("MSE score testing =", mean_squared_error(Y_test, Y_test_pred))
print("MAE score testing =", mean_absolute_error(Y_test, Y_test_pred))
#print(r2_score(Y_test,Y_test_pred))
```
